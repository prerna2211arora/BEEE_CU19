**QUESTION 5: Plot graphs of various columns to check for outliers and remove those data points from the
dataset.**


```
import seaborn as sns
sns.boxplot(data["MSRP"])   #DELETING THE OUTLIERS BEFORE TREATING TIME
```

```
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3-Q1
print(IQR)
data = data[~((data < (Q1 - 1.5*IQR)) |(data > (Q3 + 1.5*IQR))).any(axis=1)]
```

```
sns.boxplot(data["MSRP"])
```

**QUESTION 6: What car brands are the most represented in the dataset and find the average price among
the top car brands.**

```
array=data["Make"].value_counts()[:10].index.tolist()
array1=data.groupby("Make")['MSRP'].mean().to_dict()
top_car_brands={}
for i in range(0,10):
  top_car_brands[i+1]={array[i] : array1[array[i]]}
print("The average price among the top car brands are: \n")
top_car_brands
```
